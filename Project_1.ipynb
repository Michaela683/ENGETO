```python
"""
projekt_1.py: první projekt do Engeto Online Python Akademie
author: Michaela Žák Štarková
email: michaela.zakstarkova@gmail.com
discord: michaelazak_07263
"""                                                                               

# VÝBĚR UŽIVATELE
USERS_PASSWORDS = {
        "bob": "123",
        "ann": "pass123",
        "mike": "password123",
        "liz": "pass123"
        }
oddelovac = "-" *50
user_name = input("Enter your user name: ") # Vyžádá si od uživatele přihlašovací jméno a heslo
password = input("Enter your password: ")

while (user_name, password) not in USERS_PASSWORDS.items(): # zjistí, jestli zadané údaje odpovídají někomu z registrovaných uživatelů
    print(f"Username: {user_name}\nPassword: {password}\nWrong user name or password. Terminating programme.")  # pokud není registrovaný, upozorní jej a ukončí program
    break
else:
   print(f"username: {user_name}\npassword: {password}")
   print(oddelovac)
   print(f"Welcome to the app, {user_name.capitalize()}.\nWe have 3 texts to be analyzed.") # pokud je registrovaný, pozdraví jej a umožni mu analyzovat texty
   print(oddelovac)


   TEXTS = {
    "1": ['''Situated about 10 miles west of Kemmerer,
    Fossil Butte is a ruggedly impressive
    topographic feature that rises sharply
    some 1000 feet above Twin Creek Valley
    to an elevation of more than 7500 feet
    above sea level. The butte is located just
    north of US 30N and the Union Pacific Railroad,
    which traverse the valley. '''],
    "2": ['''At the base of Fossil Butte are the bright
    red, purple, yellow and gray beds of the Wasatch
    Formation. Eroded portions of these horizontal
    beds slope gradually upward from the valley floor
    and steepen abruptly. Overlying them and extending
    to the top of the butte are the much steeper
    buff-to-white beds of the Green River Formation,
    which are about 300 feet thick.'''],
    "3": ['''The monument contains 8198 acres and protects
    a portion of the largest deposit of freshwater fish
    fossils in the world. The richest fossil fish deposits
    are found in multiple limestone layers, which lie some
    100 feet below the top of the butte. The fossils
    represent several varieties of perch, as well as
    other freshwater genera and herring similar to those
    in modern oceans. Other fish such as paddlefish,
    garpike and stingray are also present.''']
    }

   oddelovac = "-" * 40

# VÝBĚR TEXTU
   print(f"Enter a number btw. 1 and 3 to select: ")

   while True:
        selection = input("Enter a number btw. 1 and 3 to select: ")

        if not selection.isdigit():  # Pokud zadání není číslice
            print("Enter is not numerical.")
            #break
        elif selection not in ("1", "2", "3"): # Pokud zadání není číslo 1 - 3
            print(f"Text number {selection} cannot be selected.")
            break
        else:
            print(f"Text selected: {selection}.") # Pokud zadání je číslo 1 - 3
            break
   print(oddelovac)


# TEXTOVÝ ANALYZÁTOR
   for text in TEXTS[selection]: 
        words = text.split() 
        num_words = len(words)
        print(f"There are {num_words} words in the selected text.")

   titles = 0
   for word in words: # Najde počet slov začínajících velkým písmenem
        if word.istitle():
            titles +=1
   print(f"There are {titles} titlecase words.")

   capitals = 0 # Najde počet slov psaných velkými písmeny
   for word in words:
        if word.isupper() and word.isalpha():
                capitals += 1
   print(f"There are {capitals} uppercase words.")

   lower_case = 0
   for word in words: # Najde počet slov psaných malými písmeny
        if word.islower():
            lower_case +=1
   print(f"There are {lower_case} lowercase words." )

   digits = 0
   for word in words: # Najde počet čísel (ne cifer)
        if word.isdigit():
            digits +=1
   print(f"There are {digits} numeric strings.")

   digits_sum = []
   for word in words: # Najde sumu všech čísel (ne cifer) v textu.
        if word.isdigit():
            digits_sum.append(int(word))
   sum(digits_sum) 
   print(f"The sum of all numbers is {sum(digits_sum)}.")

   new_text = text.replace(",","", ).replace(".", "") # Odstraní čárky a tečku v textu

   new_words = (new_text.split()) # Rozdělí nový text na slova

   word_lengths_dict = {} # dict KEYS = slova a VALUES = jejich délka
    
   for new_word in new_words:
        length = len(new_word)
        word_lengths_dict[new_word] = length

    
   unique_count = str(word_lengths_dict.values())  # délka slov převedená na string
   c = list([len(unique_count) for unique_count in word_lengths_dict]) # délky slov jako list


   tuple_c = []
   [tuple_c.append(item) for item in c if item not in tuple_c] # dlky slov jako tuple, aby se neopakovaly
   sort_c = sorted(tuple_c)


   d = {unique_count:c.count(unique_count) for unique_count in c} # d = dict, KEYS = délky slov a VALUES = jejich frekvence
   d_list = list(d.values())

   print(oddelovac)
   print(f"LEN| OCCURENCES  |NR.")
   print(oddelovac)

   for x, y in zip(sort_c, d_list):
        print(f"{x:<3}| {"*"*y:<12}| {y:<10}")
```
